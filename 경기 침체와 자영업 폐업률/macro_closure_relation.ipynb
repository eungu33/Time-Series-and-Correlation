```python
# 1. 데이터 불러오기
import pandas as pd
import openpyxl

file = 'data/소비자_물가_지수.xlsx'
DF = pd.read_excel(file)
DF = DF.T
DF.columns=['소비자물가지수','소비자물가상승률']
DF
```

```python
cpi = DF.drop('Unnamed: 0')
cpi
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(8, 5))

# 좌측 Y축 (소비자물가지수)
ax1.plot(cpi.index, cpi["소비자물가지수"], color="blue", marker="o", label="소비자물가지수")
ax1.set_ylabel("소비자물가지수", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# 우측 Y축 (소비자물가상승률)
ax2 = ax1.twinx()
ax2.plot(cpi.index, cpi["소비자물가상승률"], color="red", marker="s", linestyle="dashed", label="소비자물가상승률")
ax2.set_ylabel("소비자물가상승률 (%)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# X축 설정
ax1.set_xlabel("연도")

# 제목 추가
plt.title("연도별 소비자물가지수 및 소비자물가상승률")

# 범례 추가
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# 그래프 표시
plt.show()
```

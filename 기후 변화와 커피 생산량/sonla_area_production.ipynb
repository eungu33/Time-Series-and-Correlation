```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
coffee = pd.read_excel('data/베트남 원두 생산량.xlsx')
```

생산량과 재배면적 <hr>

```python
# 데이터 전처리
# 1. 커피 생산량 데이터 처리                  # 생산량
coffee_prod = coffee[coffee['Element'] == 'Production']
coffee_Area = coffee[coffee['Element'] == 'Area harvested']

coffee_prod = coffee_prod.set_index('Year')['Value']
coffee_Area = coffee_Area.set_index('Year')['Value']
```

```python
# 그래프 생성
plt.figure(figsize=(15, 7))

ax1 = plt.gca()  # 현재 Axes 가져오기
ax1.plot(coffee_prod.index, coffee_prod.values, 'b-', label='Production')
ax1.set_xlabel('연도')
ax1.set_ylabel('커피 생산량 (톤)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()  # 두 번째 y축
ax2.plot(coffee_Area.index, coffee_Area.values, 'r-', label='Area harvested')
ax2.set_ylabel('재배 면적 (ha)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# 범례 설정
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('베트남 커피 생산량 재배 면적 비교')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

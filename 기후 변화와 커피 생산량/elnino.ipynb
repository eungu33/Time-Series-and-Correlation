```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
elnino = pd.read_excel('data/elnino_data.xlsx')
elnino.head()
```

```python
# 연도(YR)와 월(MON)을 합쳐 날짜(datetime)로 변환 (DAY=1을 명시적으로 추가)
elnino["Date"] = pd.to_datetime(elnino['YR'].astype(str) + '-' + elnino['MON'].astype(str) + '-1')

# 불필요한 열 삭제
elnino = elnino.drop(columns=["YR", "MON"])

print(elnino)
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6),facecolor='lightblue')
plt.plot(elnino['Date'], elnino['DATA'],color='#B4E7E2', linestyle='-')

plt.axhline(0.0, color='#FFB3B3', linestyle='--', linewidth=1)

# 제목과 라벨 추가
plt.title('엘니뇨 감시구역의 해수면온도 편차', fontsize=16)
plt.xlabel('연도')
plt.ylabel('해수면 온도 편차', fontsize=12)
plt.xticks(rotation=45)
# plt.grid(True)

# plt.gca().set_facecolor('#40e0d0')
# plt.gca().set_facecolor('#A7C7E7')

# 그래프 출력
plt.tight_layout()
plt.show()
```

2020년~2023년 라니냐:
